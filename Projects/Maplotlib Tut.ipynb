# MATLAB-Stil Vs. Objektorientierte Schnittstelle

```python
plt.figure() # Abbildng erzeugen 
# Erstes Diagramm erzeugen und Achse zuweisen
plt.subplot(2,1,1)
plt.plot(x, np.sin(x))
# Zweites Diagramm erzeugen und Achse zuweisen
plt.subplot(2,1,2)
plt.plot(x, np.cos(x))
```

```python
# Raster fÃ¼r die Diagrammne erzeugen
# ax ist ein Array, das zwei Axes-Objekte enthÃ¤lt
fig, ax = plt.subplots(2)
# plot()-Methoden der Objekte aufrufen
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

# Anpassen des Diagrammsbegrenzung

```python
plt.plot(x, np.sin(x))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
```

**plt.axis()** (mit i) ist eine nÃ¼tzliche Funktion, die es ermÃ¶glicht, die Begrenzungen der x- und y-Achse mit einem einzigen Aufrug einzustellen.


```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5]) # xmin, xmax, ymin, ymax = axis([xmin, xmax, ymin, ymax])
```

Man kann auch die axis Methide verwenden um die Begrenzung automatisch eng am aktuellen Graphen auszurichten.
Mehr unter [link](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.axis.html).

```python
plt.plot(x, np.sin(x))
plt.axis('tight')
```

# Diagramme beschriften

```python
plt.plot(x, np.sin(x))
plt.title("Eine Sinuskurve")
plt.xlabel("x")
plt.ylabel("sin(x)")
```

Wenn in einem einzigen axes Objekt mehrere Kurven angezeigt werden, ist es sinnvoll, eine Legende zur Kennzeichung der vrschiedenen Liinenarten zu erstellen. Auch hierfÃ¼r bietet Marplotlib eine integriert MÃ¶glichkeit: die *plt.legend()* Methode

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')
plt.legend()
```
